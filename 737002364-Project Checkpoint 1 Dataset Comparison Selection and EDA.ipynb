# (D) Exploratory Data Analysis (Selected Dataset Only)

```python
import json
from pathlib import Path

import pandas as pd
import matplotlib.pyplot as plt

# ---------- Config ----------
TRAIN_PATH = Path(r"C:\VS CODE\Sem long\otto-recsys-train.jsonl")
N_SESSIONS = 20_000  # sampling for feasibility

# ---------- Load sample (memory-safe) ----------
sessions = []
with TRAIN_PATH.open("r", encoding="utf-8") as f:
    for i, line in enumerate(f):
        sessions.append(json.loads(line))
        if i + 1 >= N_SESSIONS:
            break

# ---------- Flatten to event-level table ----------
rows = []
for s in sessions:
    sid = s["session"]
    for e in s["events"]:
        rows.append((sid, e["aid"], e["ts"], e["type"]))

df = pd.DataFrame(rows, columns=["session", "aid", "ts", "type"])

# ---------- Basic stats ----------
n_sessions = df["session"].nunique()
n_events = len(df)
n_items = df["aid"].nunique()
event_type_counts = df["type"].value_counts()
event_type_props = df["type"].value_counts(normalize=True)

# ---------- Session length stats ----------
session_lengths = df.groupby("session").size()
session_len_desc = session_lengths.describe()

# ---------- Item popularity stats ----------
item_counts = df["aid"].value_counts()
top1_share = item_counts.head(max(1, int(0.01 * len(item_counts)))).sum() / item_counts.sum()
top10_share = item_counts.head(10).sum() / item_counts.sum()

# ---------- Temporal prep ----------
df["ts"] = pd.to_datetime(df["ts"], unit="ms")
df = df.sort_values(["session", "ts"])

# Compute per-session time span (seconds) and median gap (seconds)
tmp = df.copy()
tmp["dt_sec"] = tmp.groupby("session")["ts"].diff().dt.total_seconds()
median_gap_sec = tmp["dt_sec"].median(skipna=True)

session_span_sec = (df.groupby("session")["ts"].max() - df.groupby("session")["ts"].min()).dt.total_seconds()
span_desc = session_span_sec.describe()

# ---------- Create basket + sequence views (for later techniques) ----------
basket = df.groupby("session")["aid"].apply(lambda x: list(pd.unique(x)))   # unique items, preserves first-seen order
sequence = df.groupby("session")["aid"].apply(list)                        # full ordered sequence with repeats

# ---------- Print a compact report ----------
print("=== OTTO Sample EDA Summary ===")
print(f"Loaded sessions: {len(sessions):,}")
print(f"Event rows: {n_events:,}")
print(f"Unique items: {n_items:,}")

print("\nEvent type counts:")
print(event_type_counts.to_string())

print("\nEvent type proportions:")
print((event_type_props * 100).round(2).astype(str) + "%")

print("\nSession length (events per session) summary:")
print(session_len_desc.to_string())

print("\nSession time span (seconds) summary:")
print(span_desc.to_string())

print(f"\nMedian time gap between consecutive events (sec): {median_gap_sec:.2f}")

print("\nItem popularity concentration:")
print(f"Top 10 items share of all events: {top10_share:.4f}")
print(f"Top 1% items share of all events: {top1_share:.4f}")

# ---------- Plots (3 clean plots) ----------
plt.figure(figsize=(8, 5))
plt.hist(session_lengths, bins=50)
plt.title("Session Length Distribution (events per session)")
plt.xlabel("Events per session")
plt.ylabel("Count")
plt.show()

plt.figure(figsize=(10, 5))
item_counts.head(20).plot(kind="bar")
plt.title("Top 20 Items by Event Count")
plt.xlabel("Item (aid)")
plt.ylabel("Event Count")
plt.show()

plt.figure(figsize=(8, 5))
event_type_counts.plot(kind="bar")
plt.title("Event Type Counts")
plt.xlabel("Event Type")
plt.ylabel("Count")
plt.show()
```

## Sampling Justification

I sampled the first **20,000 sessions** from the OTTO training dataset to ensure computational feasibility on a local machine while preserving realistic behavioral patterns. 

This sample contains:

- **1,127,647 interaction events**
- **273,569 unique items**

The sample size is sufficiently large to capture real-world characteristics such as heavy-tailed session behavior and item popularity skew, while remaining manageable for exploratory analysis and algorithm experimentation.

---

## Data Quality and Structural Observations

The dataset is structured as session-level interaction logs in JSONL format. Each session contains time-ordered interaction events consisting of:

- Item ID (`aid`)
- Timestamp (`ts`)
- Event type (`clicks`, `carts`, `orders`)

The data was flattened into an event-level tabular structure and sorted chronologically within each session to enable both unordered and ordered pattern analysis.

Repeated interactions with the same item occur within sessions, which may influence support counts and require careful treatment when constructing transaction baskets.

---

## Event Type Distribution

Interaction types are highly imbalanced:

- **Clicks:** 1,027,514 (91.12%)
- **Carts:** 79,598 (7.06%)
- **Orders:** 20,535 (1.82%)

This indicates that purchase intent signals (orders) are sparse relative to browsing activity. Frequent itemset mining based purely on event counts may therefore emphasize browsing behavior rather than purchasing behavior.

---

## Session Length Characteristics

Session length distribution is strongly right-skewed:

- **Mean:** 56.38 events
- **Median:** 21 events
- **25th percentile:** 6 events
- **Minimum:** 2 events

Most sessions are short, while a small number of sessions are extremely long (heavy tail). This suggests that a single global support threshold may bias frequent pattern discovery toward high-activity sessions.

---

## Item Popularity Concentration

Item frequency analysis reveals moderate concentration:

- Top 10 items account for **0.55%** of all events.
- Top 1% of items account for **18.87%** of all events.

Although the very top items do not dominate completely, there is clear long-tail behavior. This implies that:

- High support thresholds may mostly capture popular-item co-occurrences.
- Lower support thresholds may be necessary to uncover niche patterns.

---

## Initial Observations Motivating External Techniques

The exploratory analysis reveals three important structural properties:

1. **Heavy-tailed session behavior** suggests that temporal ordering may contain meaningful behavioral structure beyond simple co-occurrence.
2. **Interaction imbalance (91% clicks)** indicates that purchase intent is sparse, motivating analysis beyond raw frequency counts.
3. **Moderate popularity concentration** implies that unordered frequent itemsets may primarily reflect exposure-driven associations rather than behavioral sequences.

Because this dataset preserves full temporal ordering of interactions, it is well-suited for comparing:

- **Unordered frequent itemsets and association rules** (course technique)
- **Sequential pattern mining** (beyond-course technique)

Sequential analysis may reveal progression patterns such as click ‚Üí cart ‚Üí order transitions that unordered baskets cannot capture.


## Key Observations from Exploratory Analysis
Based on the 20,000-session sample of the OTTO dataset, several structural properties emerge:

### 1. Heavy-Tailed Session Distribution
Session lengths are strongly right-skewed. While the median session contains 21 events, some sessions contain hundreds of interactions. This suggests heterogeneous user behavior and raises concerns about bias when applying global support thresholds.

### 2. Interaction Imbalance
Clicks represent over 91% of all interactions, while orders account for less than 2%. This indicates that purchase behavior is sparse relative to browsing activity. Mining frequent patterns without considering interaction type may primarily capture browsing co-occurrence rather than purchase intent.

### 3. Long-Tail Item Distribution
Although the top 10 items account for only 0.55% of total events, the top 1% of items account for nearly 19% of all interactions. This confirms the presence of long-tail dynamics and moderate popularity concentration.

These characteristics suggest that frequent itemsets may predominantly reflect exposure-driven associations rather than structured behavioral transitions.

---

## Hypothesis

High global support thresholds in frequent itemset mining may overlook meaningful but less frequent behavioral sequences.

Specifically:

- Unordered co-occurrence patterns may fail to capture progression behaviors.
- Temporal ordering may reveal structured interaction flows such as click ‚Üí cart ‚Üí order transitions.
- Sequential patterns may provide stronger behavioral insight than static baskets.

---
